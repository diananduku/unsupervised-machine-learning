```python
##dropping the customer_id column
df=data.drop('CUST_ID',axis=1)
df
```

```python
##visualization of individual features
import matplotlib.pyplot as plt
import seaborn as sns
fig, axes = plt.subplots(nrows=6, ncols=3, figsize=(12, 12))
variables = df.columns
axes = axes.flatten()
# Loop through each variable and create subplots
for i, variable in enumerate(variables):
    ax = axes[i]
    sns.histplot(df[variable], ax=ax, kde=True, bins=20)
    ax.set_title(f'Distribution of {variable}')
    ax.set_xlabel(variable)
    ax.set_ylabel('Frequency')
    
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])
    
plt.subplots_adjust(wspace=0.4, hspace=0.8) 
```

```python
##since the data is higly varied, normalize it
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
normalized_data=scaler.fit_transform(df)
```

```python
#column_names=data.columns
data1=pd.DataFrame(normalized_data,columns=df.columns)
data1
```

```python
##performing k-means clustering
from sklearn.cluster import KMeans
x=normalized_data
```

```python
##elbow method to calculate the optimal number of clusters
sse = []  # Initialize the list to store SSE values
k_range = range(1, 11)  # Define the range of K values

for k in k_range:
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=1)
    kmeans.fit(x)
    sse.append(kmeans.inertia_)

plt.figure(figsize=(8, 6))
plt.plot(k_range, sse, marker='o', linestyle='-', color='b')
```

```python
##ideal number of clusters is 2
```

```python
kmeans = KMeans(n_clusters=2,init='k-means++')
labels=data1['cluster_label'] = kmeans.fit_predict(x)
labels
```

```python
labels.shape
```

```python
##visualizing the clusters for balance against purchases
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data1, x='BALANCE', y='PURCHASES', hue='cluster_label', palette='viridis', s=100)
plt.title('K-means Clustering of Balance against Purchases')
plt.xlabel('BALANCE')
plt.ylabel('PURCHASES')
plt.legend(title='Cluster_label')
plt.show()
```

```python
##visualizing the clusters for payments against cash_sdvance_trx
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data1, x='PAYMENTS', y='CASH_ADVANCE_TRX', hue='cluster_label', palette='viridis', s=100)
plt.title('K-means Clustering of Payments against Cash Advance')
plt.xlabel('PAYMENTS')
plt.ylabel('CASH_ADVANCE_TRX')
plt.legend(title='Cluster_label')
plt.show()
```

```python
##since the clusters are overlapping, use PCA to make principal components
from sklearn.decomposition import PCA
```

```python
##scree plot to determine optimal number of principal components
pca = PCA()
pca.fit(df)
explained_variances = pca.explained_variance_ratio_

# Create a scree plot
plt.figure(figsize=(8, 6))
plt.bar(range(1, len(explained_variances) + 1), explained_variances, alpha=0.7, align='center')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.title('Scree Plot')
plt.show()





```

```python
##number of components is 2 ; following the elbow point
```

```python
numComponents=2
pca=PCA(n_components=numComponents)
principal_components=pca.fit_transform(data1)
principal_components
```

```python
##k-means clustering for the principal components
cluster_labels = kmeans.fit_predict(principal_components)
cluster_labels
```

```python
# Visualize the results
plt.scatter(principal_components[:, 0], principal_components[:, 1], c=cluster_labels)
plt.title('K-means Clustering with 2 Principal Components')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()

```

```python
##silhouette score to determine quality of clustering
from sklearn.metrics import silhouette_score
silhouette_score = silhouette_score(principal_components, cluster_labels)
silhouette_score
```

```python
##since silhouette test is closer to 1, this indicated batter quality of clustering
```

```python
#davies_bouldin_score
from sklearn.metrics import davies_bouldin_score
db_index = davies_bouldin_score(principal_components, cluster_labels)
db_index
```

```python
##the score is lower indicating better clustering
```
